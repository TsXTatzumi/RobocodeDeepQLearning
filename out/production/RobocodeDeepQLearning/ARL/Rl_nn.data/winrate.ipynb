```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerLine2D
#from scipy.interpolate import spline
#from pandas import Series
#import pandas as pd

winValue = []
numRounds = []
winRates = []

winValueFile = open("winrate.txt", "r") 
winValue =  winValueFile.readlines()
rounds = len(winValue)
numRounds = np.arange(0,rounds)
valueSum = 0

for i in range(0, rounds):
    valueSum += int(winValue[i])
    winRates.append(valueSum / (i+1))  


x = np.array(numRounds)
y = np.array(winRates)


axes = plt.gca()
axes.set_xlim([0,rounds])
axes.set_ylim([0,1])
plt.plot(x, y,linewidth=2.0) 
plt.xlabel('No. of Rounds')
plt.ylabel('Win rate')
plt.title('Win rate vs No. of Rounds')
plt.show()
```
